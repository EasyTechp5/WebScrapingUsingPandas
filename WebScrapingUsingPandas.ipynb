```python
link2 = 'https://www.theguardian.com/football/tables'
```

```python
data2 = pd.read_html(link2)
```

```python
data2[0]
```

```python
data2[0].iloc[:4,:]
```

```python
extracted_data = [x.iloc[:4,:] for x in data2]
extracted_data[0]
```

```python
extracted_data[1]
```

```python
# Using the for loop
extractdata2 = []

for table in data2:
  extract = table.iloc[:4,:]
  extractdata2.append(extract)
```
